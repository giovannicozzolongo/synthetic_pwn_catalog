## Models for pulsar wind nebula evolution

### Baseline model
Use this baseline model taken from the pwn population paper by the HESS collaboration, A&A 612 (2018) A2.

```python
# Read in .dat file and define an interpolator. 
baselinemodel = "../data/pwnpopbaselinemodel.dat"
t_psr, tau_psr, Edot_psr, B_pwn, R_pwn, L_pwn, ind_pwn = np.genfromtxt(baselinemodel).T

# Properties of the PWN can be obtained provided the Edot (spin-down luminosity) is known for the synthetic population.
edot = 10.**np.linspace(np.log10(1e33),np.log10(1e38),50)

# Interpolation
t_interpolator = interp1d(Edot_psr, t_psr, bounds_error=False, fill_value=(t_psr[0], t_psr[-1]))
R_interpolator = interp1d(Edot_psr, R_pwn, bounds_error=False, fill_value=(R_pwn[0], R_pwn[-1]))
L_interpolator = interp1d(Edot_psr, L_pwn, bounds_error=False, fill_value=(L_pwn[0], L_pwn[-1]))
ind_interpolator = interp1d(Edot_psr, ind_pwn, bounds_error=False, fill_value=(ind_pwn[0], ind_pwn[-1]))

times = t_interpolator(edot)
radii = R_interpolator(edot)
lum1to10 = L_interpolator(edot)
indices = ind_interpolator(edot)

tau_0 = 3. # [kyr]
init_edot = edot*(1.+times/tau_0)**2.

# scale luminosities and radii
lum1to10 *= init_edot / 1.e39 # baseline initial edot
radii *= (init_edot / 1.e39)**0.2
```

### Varied model
Varied model ranges from Table A.1.

```python
# Compute the magnetic field time evolution
def E_dot(t):
    n = 3 # Braking index
    E_dot = E_dot_0 * (1 + (t / tau_0) ** (-(n + 1) / (n - 1)))
    return E_dot

# Compute the magnetic field time evolution
alpha = 0.6 # In order to satisfy the conservation of magnetic flux
B_ISM = 3 # muG
def B(t):
    B = B_0 / (1 + (t / tau_0) ** alpha) + B_ISM
    return B

# Compute the PWN radius time evolution
def R(t):
    if (t_rs >= tau_0):
        if (t <= tau_0):
            R = t ** (6/5)
        elif ((tau_0 < t) & (t <= t_rs)):
            R = t
        else:
            R = t ** (3/10)
    else:
        if (t <= t_sr):
            R = t ** (6/5)
        elif ((t_sr < t) & (t <= tau_0)):
            R = t ** (11/15)
        else:
            R = t ** (3/10)
    return R

# Compute the TeV surface brightness
def S(df):
    L = df['E_dot'] ** 0.58 # TeV luminosity
    S = L / (4 * np.pi * df['R'] ** 2)
    return S
```

```python
# Choose randoms values withing the varied model ranges
E_dot_0 = uniform(2.5, 3.5) # Initial spin-down power [E39 erg/s]
tau_0 = uniform(1, 4) # Initial spin-down timescale [kyr]
B_0 = uniform(110, 270) # Initial magnetic field strength [muG]
t_rs = uniform(4, 8) # Reverse shock interaction timescale [kyr]
R_3 = uniform(3, 12) # PWN radius at t = 3 kyr [pc]

# Create a dataframe for the varied model
df = pd.DataFrame()
df['E_dot'] = pd.Series(t_psr).apply(E_dot)
df['B_ISM'] = pd.Series(t_psr).apply(B)
df['R'] = pd.Series(t_psr).apply(R)
df['S'] = df[['E_dot', 'R']].apply(S, axis = 1)

# Results
df
```

```python
plt.plot(df['E_dot'], df['R'])
plt.semilogx()
plt.semilogy()
plt.ylabel("Gamma-ray luminosity 1-10 TeV")
plt.xlabel("Gamma-ray spectral index")
```

## Read the catalog CAMPAS
For the stellar type columns, NSs are identified as 13. 14 are BHs, 10-13 are WDs (He, CO, ONe). 15 is a massless remnant (SN left no remnant). Numbers below 11 identify giant and MS stages. These codes follow Hurley et al. (2000).

```python
COMPAS_OUTPUT_FOLDER = Path('../data/2023_21XX_pulsar_proc_compas_output/COMPAS_1e2FeH=30.snappy.parquet')

# Load the entire sample, which includes *all* systems passed to '''COMPAS''' as an initial population,
# both ones that led to pulsar formation and ones that did not.
sample_df = pd.read_parquet(path=COMPAS_OUTPUT_FOLDER,
                           engine='pyarrow',
                           use_threads=True)

# Load only pulsar-hosting systems
psr1_df = pd.read_parquet(path=COMPAS_OUTPUT_FOLDER,
                          engine='pyarrow',
                          filters=[('PSR_Type1', '==', 13.)],
                          use_threads=True)

psr2_df = pd.read_parquet(path=COMPAS_OUTPUT_FOLDER,
                          engine='pyarrow',
                          filters=[('PSR_Type2', '==', 13.)],
                          use_threads=True)
```

We need just a few operations to recover $\dot{E}$, using

$$\dot{E}=4\pi^2 I\frac{\dot{P}}{P}.$$

```python
ns_moment_inertia = 1e45 # gm cm2
def edot1(row):
    p = row.PSR_P1
    pdot = row.PSR_Pdot1
    edot1 = 4 * np.pi * np.pi * ns_moment_inertia * pdot / p
    return edot1
```

```python
psr1_df = psr1_df[psr1_df.PSR_Omega1 > 0]
```

```python
psr1_df.PSR_Pdot1 = np.abs(psr1_df.PSR_Pdot1)
```

```python
psr1_df['PSR_P1'] = 2*np.pi/psr1_df.PSR_Omega1
```

For now I had to implement the filter below in order to eliminate some completely aphysical systems (you can generate the graphs without apply the filter and see what I mean). I am not sure why these are showing up in the sample, but, looking at the COMPAS pulsar paper, they are likely not supposed to be there. We can keep manually removing them from the sample until I figure out where they are coming from.

```python
psr1_df = psr1_df[(psr1_df.PSR_P1 > 5e-3) & (psr1_df.PSR_P1 < 1e10)]
```

```python
# Add period, period derivative and magnetic field
psr1_df['PSR_logP1'] = np.log10(psr1_df.PSR_P1)
psr1_df['PSR_logPdot1'] = np.log10(psr1_df.PSR_Pdot1)
psr1_df['PSR_logB1'] = np.log10(psr1_df.PSR_B1)

# View the columns 
psr1_df.columns
```

```python
# Generate luminosity values
psr1_df['PSR_Edot1'] = psr1_df.apply(edot1, axis=1)

# Keep only luminosity values within the baseline ones
#psr1_df['PSR_Edot1'] = psr1_df[(psr1_df['PSR_Edot1'] >= np.min(Edot_psr)) & ( psr1_df['PSR_Edot1'] <= np.max(Edot_psr))]['PSR_Edot1']
psr1_df = psr1_df[(psr1_df['PSR_Edot1'] >= np.min(Edot_psr)) & ( psr1_df['PSR_Edot1'] <= np.max(Edot_psr))]

# Generate the gamma ray index
psr1_df['index'] = ind_interpolator(psr1_df['PSR_Edot1'])

psr1_df
```
